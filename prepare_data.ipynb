```python
import csv
import pandas as pd
import numpy as np
```

### Train dataset

```python
with open('train', 'r') as f:
    strip_line = (line.strip('\n') for line in f)
    lines = (line.split(' ') for line in strip_line if line)
        
    with open('train.csv', 'w') as outfile:
        writer = csv.writer(outfile)
        writer.writerow(('sentence_idx', 'word', 'NER_tag'))
        writer.writerows(lines)
```

```python
df = pd.read_csv('train.csv')
train = df.drop(df[df['word']=='-DOCSTART-'].index)
train['word'].fillna('NA', inplace=True)
```

```python
indices = train.loc[train['sentence_idx'] == 1].index.tolist()
```

```python
df = pd.DataFrame({'sentence_idx':0,'word':np.nan,'NER_tag':np.nan}, index=[i-0.5 for i in indices])
```

```python
df
```

```python
df_new = train.append(df)
```

```python
df_new = df_new.sort_index()
```

```python
df_new = df_new.reset_index(drop=True)
```

```python
df_new
```

```python
df_new = df_new.drop('sentence_idx', axis=1)
```

```python
df_new = df_new.drop(0, axis=0)
df_new
```

```python
df_new.to_csv('train.tsv', sep = '\t', index=None, header=None)
```
